```python
#Clean Data using Pandas

import pandas as pd

df=pd.read_csv("dirtydata.csv")
print(df.head())

```

```python
df["Date"] = pd.to_datetime(df["Date"])
print(df)
```

```python
#drop the NONE elements in the Spesific

df.dropna(subset=["Date"],inplace=True)
print(df)
```

```python
df.loc[7,"Duration"]=120
print(df)
```

```python
for i in df.index:
    if df.loc[i,"Duration"] > 100 :
        df.loc[i,"Duration"]=130
print(df)
```

```python
#Duration Cloumn filter big range 


for i in df.index:
    if df.loc[i,"Duration"] >100:
        df.drop(i,inplace=True)
print(df)
```

```python
#find the duplecates


import pandas as pd


df=pd.read_csv("dirtydata.csv")

x=df.duplicated()

print(x)
```

```python
#REmove the duplicates


df=pd.read_csv("dirtydata.csv")

df.drop_duplicates(inplace=True)

print(df)
```

```python
#its helps to the data is use full or not use

#the Score is -1 then the data is useless and 1 then data is perfect

print(df.corr())
```
